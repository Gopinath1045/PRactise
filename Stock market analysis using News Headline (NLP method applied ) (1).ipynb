# teXt pre-processing

```python
#convert all the text into vectors , so that we can apply ML algorithm
```

```python
#need to combine the each of the column then cobmine all the data as paragraph
```

```python
#each row combined
' '.join(str(x) for x in data.iloc[1,0:25])
```

```python
#combining all the records of the data
headlines=[]
for row in range(0,len(data.index)):
    headlines.append(' '.join(str(x) for x in data.iloc[1,0:25]))
```

```python
headlines[2]
```

```python
'''#combining all the records of the data
#each row combined
' '.join(str(x) for x in data_test.iloc[1,0:25])
headlines_test=[]
for row in range(0,len(data_test.index)):
    headlines_test.append(' '.join(str(x) for x in data_test.iloc[1,0:25]))'''
```

```python
#now we will convert text into vectors
```

```python
#count vectorizer
#Convert a collection of text documents to a matrix of token counts.
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
##implement bag of words
countvector=CountVectorizer(ngram_range=(2,2))
traindataset=countvector.fit_transform(headlines)
```

```python
traindataset
```

```python
traindataset[1]
```

```python
#using random forest classifier
```

```python
 from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier(n_estimators=200,criterion='entropy')
rf.fit(traindataset,train['Label'])
```

```python
## Predict for the Test Dataset
test_headline= []
for row in range(0,len(test.index)):
    test_headline.append(' '.join(str(x) for x in test.iloc[row,2:27]))
test_dataset = countvector.transform(test_headline)
predictions = rf.predict(test_dataset)
```

```python
predictions
```

```python
## Import library to check accuracy
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
matrix=confusion_matrix(test['Label'],predictions)
print(matrix)
score=accuracy_score(test['Label'],predictions)
print(score)
report=classification_report(test['Label'],predictions)
print(report)
```

```python
sns.heatmap(matrix,yticklabels=True, annot=True,fmt='d')
```
